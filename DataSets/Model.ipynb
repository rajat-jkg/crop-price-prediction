```python
import pandas as pd
import matplotlib.pyplot as mlp
from sklearn import linear_model
```

```python
df = pd.read_excel('final-data.xlsx')
df.head()
```

```python
%matplotlib inline

mlp.scatter(df['rain_sum_jan_to_may_sum'], df['Yield_hg_ha'])
```

```python
df = df.query('rain_sum_jan_to_may_sum>3500')
```

```python
%matplotlib inline
mlp.xlabel('Rainfall form Jan to April')
mlp.ylabel('Yield hg/ha')
mlp.scatter(df['rain_sum_jan_to_may_sum'], df['Yield_hg_ha'], marker='+', color='brown')
```

```python
model = linear_model.LinearRegression()
model.fit(df[['rain_sum_jan_to_may_sum']], df.Yield_hg_ha)
```

```python
testing_data = df.tail(3)
testing_data['rain_sum_jan_to_may_sum']
```

```python
result = model.predict(testing_data[['rain_sum_jan_to_may_sum']]).round(2)
result
```

```python
comp = testing_data
comp['predicted value'] = result
comp
```
